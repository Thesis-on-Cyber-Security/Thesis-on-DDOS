```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from dask import dataframe as dd
from sklearn.utils import shuffle
```

```python
def readCSV_Function(dir_path,file_name):
    dask_df = dd.read_csv(dir_path+file_name, blocksize=1e6)
    dask_df = dask_df.repartition(npartitions=8)
    return dask_df.compute(scheduler='threads')

filePath = "D:\\Thesis_Group_of_Brig_Gen_Razzak_Sir\\After_Column_Data_Cleaning\\"
imgPath = "G:\\Brig_Gen_Razzak_Sir_Thesis_Group\\Thesis-on-DDOS-main\\images\\"
file_df = "After_Label_Encoding.csv"
file_encoded = "Label_Encoding.csv"

Label_visualize_df = readCSV_Function(dir_path=filePath,file_name=file_encoded)
df = readCSV_Function(dir_path=filePath,file_name=file_df)
df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')
df = shuffle(df)
```

Now need to analyze the data distribution on the target feature
</center></h4>

```python
imgName = "DrDoS_LDAP Distribution.png"
encoded_value = Label_visualize_df[Label_visualize_df['Label']=='DrDoS_LDAP']['Encoding'].values[0]
((pd.DataFrame(np.log(df.where(df["Label"]==encoded_value)))).replace(-np.inf,0)).hist(figsize=(25,25),ec="white", grid=False,bins=50)
plt.savefig(imgPath+imgName)
```
